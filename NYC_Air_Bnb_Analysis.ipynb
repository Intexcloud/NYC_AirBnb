## Cleaning the data


```python
df.isnull().sum()

## delete null columns
df.dropna(inplace=True)

#check nulls
df.isnull().sum()
```

```python
df.shape
```

```python
# check duplicate
df.duplicated().sum()
```

```python
#clean duplicated
df.drop_duplicates(inplace=True)
```

```python
# check duplicate
df[df.duplicated()]
```

```python
df.dtypes
```

```python
df ['id'] = df['id'].astype(object)
```

## Data Cleaning

```python
# check outliers

data = df[df['price']<1500]
sns.boxplot(data=data, x='price')
```

```python
# Price Distribution
plt.title('price distribution')
sns.histplot(data=data, x='price',bins = 100)
plt.ylabel('Frequency')
plt.show()
```

```python
data.info()
```

```python
# Availability
plt.figure(figsize=(6,3))
plt.title('Availability')
sns.histplot(data=data, x='availability_365')
plt.ylabel('Frequency')
plt.show()
```

```python
data.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
```

```python
## price per bed
data['price per beds'] =  data['price'] / data['beds']
data.head(5)
```

```python
## average price per beds
data.groupby('neighbourhood_group')['price per beds'].mean().sort_values(ascending=False)
```

## Visualize the data

```python
## visualize by price per beds
sns.barplot(data=data, x='neighbourhood_group', y='price per beds', hue='room_type')
```

```python
## scatter plot by number of review and price
plt.title('Local review by price')
sns.scatterplot(data=data, x='number_of_reviews', y='price', hue='neighbourhood_group')
```

```python
# room type by location
sns.scatterplot(data=data, x='longitude', y='latitude', hue='room_type')
```
